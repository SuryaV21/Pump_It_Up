```python
dropp = ['date_recorded','installer','longitude','latitude','wpt_name','num_private','region_code','district_code','lga','permit','public_meeting'
         ,'recorded_by','construction_year','extraction_type_group','extraction_type_class','management_group'
         ,'payment','quantity','source_type','source_class','waterpoint_type_group']

cat_columns = ['funder','basin','subvillage','region','ward','publicpermit','scheme_management','extraction_type'
               ,'management','payment_type','water_quality','quality_group','quantity_group','source','waterpoint_type']
cont_columns = ['amount_tsh','gps_height','population','years_cons','gps_height']
```

```python
def date_to_diff(x):
    x = int(x.split('-')[0])
    if x == 0:
        return 0
    else:
        return 2020 - x


def date_to_diff2(x):
    if x == 0:
        return 0
    else:
        return 2020 - x
    
def preproc_train(comb, dropp):
    comb['years_cons'] = comb['construction_year'].apply(date_to_diff2)
    #comb = comb.drop('construction_year', axis=1)
    comb['years'] = comb['date_recorded'].apply(date_to_diff)
    comb = comb.drop(dropp, axis=1)
    #comb = comb.drop('date_recorded', axis=1)
    comb.replace('#na#', np.nan)
    return comb    
```

```python
data_m = preproc_train(data_m, dropp)
data_m.head()
```

```python
data_m = data_m.replace(np.nan, '')
```
